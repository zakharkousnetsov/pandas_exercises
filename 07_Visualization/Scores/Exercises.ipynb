### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

### Step 2. Create the DataFrame that should look like the one below.

```python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

df = pd.DataFrame(raw_data)

df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
df.plot(x='preTestScore', y='postTestScore', kind='scatter', legend=True);
# plt.legend(df['age']);
# plt.legend(["USA", "BRAZIL", "GERMANY", "INDIA", "UK"]);
# plt.legend(['A Label', 'B Label', 'C Label', 'D Label', 'E label'])
```

```python
sns.scatterplot(data=df, x='preTestScore', y='postTestScore', size='age');
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sns.scatterplot(data=df,
                x='preTestScore',
                y='postTestScore',
                hue='female',
                size=df['age'],
                sizes=(55, 222)
                );
```

```python
plt.scatter(x=df['preTestScore'],
            y=df['postTestScore'],
            c=df['female'],
            cmap='coolwarm',
            s=4.5*df['postTestScore']);
```
